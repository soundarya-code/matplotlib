```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,6,8]
y=[9,10,11,12,13,14]
# plt.plot(x,y,label='linear Graph',color='red',linewidth=2,linestyle='--',marker='.',markersize=12,markeredgecolor='blue')
# shorthand method
# fmt='[color][marker][line]'
fmt='r.--'
x2=np.arange(0,10)
# plt.plot(x,y,fmt,label='linear Graph',)
plt.plot(x[:5],x2[:5]**2,fmt)
# plt.plot(x2**2)
plt.title('first graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis of square value')
plt.xticks([0,1,2,3,4,6,8])
plt.yticks([9,10,11,12,13,14])
plt.legend()
plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]
plt.figure(figsize=(5,8),dpi=50)
bars=plt.bar(labels,values)
# bars[0].set_hatch('O')
pattern=['0','*','/']
for bar in bars:
    bar.set_hatch(pattern.pop(0))
plt.show()
```

# #piecharts


```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
plt.figure(figsize=(8,5))
labels=['left','right']
color=['#fc9d03','#805003']
plt.pie([left, right], labels = labels, colors=color, autopct='%.2f %%')
plt.title('foot Prefrences of fifa')
plt.show()
```

```python
plt.figure(figsize=(8,5))
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
```

```python
light=fifa.loc[fifa.Weight<125].count()[0]
light
```

```python
plt.figure(figsize=(8,5),dpi=300)
plt.style.use()
medium_light=fifa.loc[(fifa.Weight>=125 ) & (fifa.Weight<=150)].count()[0]
medium=fifa.loc[(fifa.Weight<=150) & (fifa.Weight<=175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight<=175)&(fifa.Weight<200)].count()[0]
heavy=fifa.loc[(fifa.Weight>=200)].count()[0]
weight=[light,medium_light,medium,medium_heavy,heavy]
# label=['<125','>=125 & <=150','<=175 & <200','>=200']
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
plt.title('weight professional scores players')
# plt.pie(weight,labels=label,pctdistance='0.8',autopct='%%.2%')
explode = (.4,.2,0,0,.4)
plt.pie(weight, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
